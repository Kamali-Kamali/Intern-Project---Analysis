```python
plt.bar(top_3.index,top_3.values)
plt.xlabel('Count')
plt.ylabel('Cuisines')
plt.title('Top 3 Cuisines')
plt.legend()
plt.show()
```

```python
plt.pie(top_3.values,labels=top_3.index,autopct='%0.2f%%')
plt.title('Top 3 Cuisines')
plt.show()
```
