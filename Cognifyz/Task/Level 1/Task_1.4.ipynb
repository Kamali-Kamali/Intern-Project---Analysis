## Determine the percentage of restaurants that offer online delivery.

## Compare the average ratings of restaurants with and without online delivery.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/Task/Cognifyz/Dataset .csv")
```

```python
df.columns
```

```python
total_count=df['Has Online delivery'].value_counts()
total_restaurant=len(df)
per=(total_count/total_restaurant)*100
percentage=per.reset_index()
percentage.columns=['Online Delivery','Percentage']
print(percentage)
```

```python
total_count=df['Has Online delivery'].value_counts()
total_restaurant=len(df)
per=(total_count['Yes']/total_restaurant)*100
print('Percentage of restaurants that offer online delivery:',per)
```

```python
compare=df.groupby('Has Online delivery')['Aggregate rating'].mean()
print('Average ratings of restaurants with and without online delivery:')
compare
```

```python
compare.plot(kind='bar')
plt.grid(axis='y',alpha=0.4)
plt.title('Ratings with and without online delivery')
plt.xlabel('Has Online Delivery')
plt.ylabel('Aggregate Rating')
plt.show()
```
