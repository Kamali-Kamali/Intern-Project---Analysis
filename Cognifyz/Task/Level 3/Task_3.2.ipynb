##Identify the restaurants with the highest and lowest number of votes.

##Analyze if there is a correlation between the number of votes and the rating of a restaurant.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/Task/Cognifyz/Dataset .csv")
```

```python
df['Votes'].max()
```

```python
high_vote=df.loc[df['Votes'].idxmax()]
print("The restaurant with the highest number of votes is:")
high_vote[['Restaurant Name','Votes','Aggregate rating']]
```

```python
print("The restaurant with the Lowest number of votes is:")
low_vote=df.loc[df['Votes'].idxmin()]
low_vote[['Restaurant Name','Votes']]
```

```python
vot=df.groupby('Restaurant Name')['Votes'].sum().sort_values(ascending=False)
```

```python
print("All over votes of restaurant from various places")
vot.head(10)
```

```python
corr=df['Votes'].corr(df['Aggregate rating'])
corr
```

```python
sns.scatterplot(x='Votes', y='Aggregate rating', data=df, alpha=0.6, color='blue')
plt.title(f"Correlation Between Votes and Aggregate Rating (r = {corr:.2f})", fontsize=14)
plt.show()
```
