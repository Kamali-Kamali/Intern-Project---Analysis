##Determine if certain cuisine combinations tend to have higher ratings.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/drive/MyDrive/Task/Cognifyz/Dataset .csv")
```

```python
df.head()
```

```python
df.columns
```

```python
cuisines_count=df['Cuisines'].value_counts()
```

```python
cuisines_count.head(10)
```

```python
cuisines_count[:20].plot(kind='barh',figsize=(10,6))
plt.xlabel('Frequence count')
plt.ylabel('Cuisines Combination')
plt.title("Top 20 Common Combinations of Cuisines")
plt.show()
```

```python
cuisine_ratings=df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False)
```

```python
cuisines_max_count=cuisines_count[:20]
```

```python
max_rating=cuisine_ratings.loc[cuisines_max_count.index]
print("Top 20 Cuisines Combinations and their Ratings:")
cuisines_max_count

max_rating
```
